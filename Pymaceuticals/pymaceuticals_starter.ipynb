```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
trial_data_complete = pd.merge( clinical_trial_data, mouse_drug_data, how='left', on='Mouse ID')

# Display the data table for preview

trial_data_complete.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_tumor=trial_data_complete.groupby(["Drug", "Timepoint"])
grouped_tumor.head()
mean_tumor_volume = grouped_tumor ['Tumor Volume (mm3)'].mean()
mean_tumor_volume.head()

# Convert to DataFrame
tumor_response_treatment = pd.DataFrame (mean_tumor_volume).reset_index()


# Preview DataFrame
tumor_response_treatment.head()

```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
error_tumor_volume = grouped_tumor ['Tumor Volume (mm3)'].sem()
# Convert to DataFrame
error_tumor_volume = pd.DataFrame (error_tumor_volume).reset_index()
# Preview DataFrame
error_tumor_volume.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

munging_error_tumor_volume =  error_tumor_volume.pivot(index = 'Timepoint', columns ='Drug', values = 'Tumor Volume (mm3)') 

# Preview that Reformatting worked

munging_error_tumor_volume
```

```python
# Minor Data Munging to Re-Format the Data Frames
munging_tumor_response_treatment =  tumor_response_treatment.pivot(index = 'Timepoint', columns ='Drug', values = 'Tumor Volume (mm3)') 
# Preview that Reformatting worked
munging_tumor_response_treatment
```

```python
# Generate the Plot (with Error Bars)

time_x = tumor_response_treatment['Timepoint'].unique()
time_x
```

```python
# Generate the Plot (with Error Bars)

fig, ax = plt.subplots()
ax.errorbar(time_x, munging_tumor_response_treatment['Capomulin'], yerr=munging_error_tumor_volume['Capomulin'], fmt="o:", color="r", label="Capomulin")
ax.errorbar(time_x, munging_tumor_response_treatment['Infubinol'], yerr=munging_error_tumor_volume['Infubinol'], fmt="^:", color="b", label="Infubinol")
ax.errorbar(time_x, munging_tumor_response_treatment['Ketapril'], yerr=munging_error_tumor_volume['Ketapril'], fmt="s:", color="g", label="Ketapril")
ax.errorbar(time_x, munging_tumor_response_treatment['Placebo'], yerr=munging_error_tumor_volume['Placebo'], fmt="d:", color="k", label="Placebo")



fig.suptitle("Tumor Response to Treatment", fontsize=16, fontweight="bold")
ax.set_xlabel("Time")
ax.set_ylabel("Tumor Volume (mm3)")
plt.grid(axis='y')

plt.legend(loc="best", fontsize="small", fancybox=True)
plt.savefig("../Images/Tumor Response to Treatment.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Minor Data Munging to Re-Format the Data Frames
munging_error_tumor_metastatic =  error_tumor_metastatic.pivot(index = 'Timepoint', columns ='Drug', values = 'Metastatic Sites') 
# Preview that Reformatting worked
munging_error_tumor_metastatic
```

```python

# Minor Data Munging to Re-Format the Data Frames
munging_metastatic_response_treatment =  metastatic_response_treatment.pivot(index = 'Timepoint', columns ='Drug', values = 'Metastatic Sites') 
# Preview that Reformatting worked
munging_metastatic_response_treatment

```

```python
# Generate the Plot (with Error Bars)
# Save the Figure
# Show the Figure

fig, ax = plt.subplots()
ax.errorbar(time_x, munging_metastatic_response_treatment['Capomulin'] , yerr=munging_error_tumor_metastatic ['Capomulin'], fmt="o:", color="r", label="Capomulin")
ax.errorbar(time_x, munging_metastatic_response_treatment['Infubinol'], yerr=munging_error_tumor_metastatic ['Infubinol'], fmt="^:", color="b", label="Infubinol")
ax.errorbar(time_x, munging_metastatic_response_treatment['Ketapril'], yerr=munging_error_tumor_metastatic ['Ketapril'], fmt="s:", color="g", label="Ketapril")
ax.errorbar(time_x, munging_metastatic_response_treatment['Placebo'], yerr=munging_error_tumor_metastatic ['Placebo'], fmt="d:", color="k", label="Placebo")


fig.suptitle("Metastatic Spread During Treatment", fontsize=16, fontweight="bold")
ax.set_xlabel("Treatment Duration (Days)")
ax.set_ylabel("Met. Sites")
plt.grid(axis='y')

plt.legend(loc="best", fontsize="small", fancybox=True)

plt.savefig("../Images/Metastatic Response to Treatment.png")
```
